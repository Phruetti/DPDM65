````python
unit_df_apart = unit_df[unit_df['propertytype_name_en'] == 'Apartment'].fillna({'price_min':np.round(unit_df[unit_df['propertytype_name_en'] == 'Apartment']['price_min'].mean())})
````

````python
unit_df_apart.isnull().any()
````

````python
unit_df_condo = unit_df[unit_df['propertytype_name_en'] == 'Condo'].fillna({'price_min':np.round(unit_df[unit_df['propertytype_name_en'] == 'Condo']['price_min'].mean())})
````

````python
unit_df_condo.isnull().any()
````

ต่อตารางแนว Y

````python
fused_df = pd.concat([unit_df_apart, unit_df_condo])
print(fused_df.shape)
fused_df.head()
````
