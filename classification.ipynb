```python
col_names=['sex','Length','Diam','Height',	'Whole'	,'Shucked',	'Viscera',	'Shell','class_age']
abalone_df = pd.read_csv('/content/drive/MyDrive/DPDM65/abalone.data',header=None,names=col_names) #แบบที่ 2
abalone_df
```

```python
from matplotlib import pyplot as plt
```

```python
plt.hist(abalone_df['class_age'],bins=len(set(abalone_df['class_age'])));
```

```python
abalone_df['CLASS']=abalone_df['class_age']>=10 #มากกว่าเท่ากับ 10
abalone_df
```

```python
abalone_df_pr=pd.get_dummies(abalone_df,columns=['sex']) #one-hot-encoder(dummy)
abalone_df_pr
```

# เริ่มทำ data mining

```python
import pandas as pd 
from sklearn.model_selection import train_test_split
```

```python
TRAIN = abalone_df_pr.iloc[:3133,:]
TEST = abalone_df_pr.iloc[3133:,:]
print(f'train={TRAIN.shape[0]},test={TEST.shape[0]}')
```

```python
x_train=TRAIN.drop(columns=['class_age','CLASS'])
x_train
```

```python
y_train=TRAIN.iloc[:,8]
y_train
```

```python
x_test=TEST.drop(columns=['class_age','CLASS'])
y_test=TEST.iloc[:,8]
```

# Decision Tree

#import

#define

#build model (fit)




```python
from sklearn.tree import DecisionTreeClassifier
```

```python
Dree1=DecisionTreeClassifier(random_state=0)
```

```python
Dree1.fit(x_train,y_train)
```

```python
from sklearn import tree
tree.plot_tree(Dree1);
```

# test

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,Dree1.predict(x_test))
```
